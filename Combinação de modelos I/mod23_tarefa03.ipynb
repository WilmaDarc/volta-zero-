# Tarefa 03

## **1.** Quais são os hyperparâmetros do RF?

> Os hyperparâmetros do algoritmo ***Random Forest*** são:
>
> - n_estimators
> - criterion
> - max_depth
> - min_samples_split
> - min_samples_leaf
> - min_weight_fraction_leaf
> - max_features
> - max_leaf_nodes
> - min_impurity_decrease
> - bootstrap
> - oob_score
> - n_jobs
> - random_state
> - verbose
> - warm_start
> - class_weight
> - ccp_alpha
> - max_samples

## **2.** Pra que serve cada um deles?

> - ***n_estimators:*** Número de árvores de decisão no modelo. Cada árvore é construída a partir de uma amostra aleatória com reposição (*Bootstrap*);
> - ***criterion:*** Métrica para avaliar a qualidade das divisões nos nós da árvore;
> - ***max_depth:*** Número para limitar a profundidade máxima da árvore;
> - ***min_samples_split:*** Número mínimo de amostras para que ocorra a divisão de um nó na árvore;
> - ***min_samples_leaf:*** Número mínimo de amostras em uma folha para que seja considerada válida na divisão;
> - ***min_weight_fraction_leaf:*** Número decimal para definir valor o "peso" (importância) mínimo necessário de uma folha;
> - ***max_features:*** Número máximo de variáveis (*feature selection*) ao buscar a melhor divisão;
> - ***max_leaf_nodes:*** Número máximo de nós na árvore;
> - ***min_impurity_decrease:*** Valor mínimo de redução da impureza que define se haverá divisão do nó de acordo com a métrica utilizada;
> - ***bootstrap:*** Define se o método *Bootstrap* será utilizado;
> - ***oob_score:*** Define se o método *out-of-bag* será utilizado caso *Bootstrap* também estiver em uso;
> - ***n_jobs:*** Número de processos que podem trabalhar em paralelo caso haja processamento disponível;
> - ***random_state:*** Controla a aleatoriedade dos métodos *Bootstrap* e *feature selection* possibilitando a reprodução dos mesmos resultados do modelo;
> - ***verbose:*** Controla o detalhamento das mensagens nos processos de treino e previsão do modelo;
> - ***warm_start:*** Possibilita o reuso do treino anterior adicionando mais árvores ao modelo;
> - ***class_weight:*** Define o peso de cada classe, possibilitando equilíbrio proporcional entre elas;
> - ***ccp_alpha:*** Parâmetro de complexidade usado para controlar a pós-poda da árvore. Quanto maior o valor, maior será a poda;
> - ***max_samples:*** Número máximo de amostras para treinar o modelo caso o método *Bootstrap* seja utilizado.

## **3.** Quais são os mais utilizados?

> Segundo pesquisas na internet, os hiperparâmetros mais utilizados e impactantes para o ajuste de um modelo Random Forest em 2025 são focados em equilibrar a capacidade preditiva com a prevenção do overfitting.
>
> **Abaixo estão os cinco principais, em ordem de importância prática**:
>1. n_estimators (Número de Árvores)
É o parâmetro mais comum. Define quantas árvores de decisão comporão a floresta. 
Uso: O padrão costuma ser 100. Valores maiores geralmente aumentam a precisão e estabilidade, mas elevam o custo computacional e o tempo de treino.
>
>Dica: Teste valores como 100, 200 ou 500 para verificar onde a taxa de erro se estabiliza
>
> 2. max_depth (Profundidade Máxima) 
Controla o crescimento vertical de cada árvore. 
Uso: É essencial para evitar o overfitting. Árvores muito profundas decoram os dados de treino; árvores rasas podem não aprender padrões complexos.
>
>Dica: Em conjuntos de dados grandes, experimente limitar entre 10 e 30 níveis.
>
> 3. max_features (Máximo de Colunas)
Define o número de características a serem consideradas em cada divisão de nó. 
Uso: É o hiperparâmetro que garante a "aleatoriedade" e diversidade da floresta.
>
>Dica: Os valores mais utilizados são "sqrt" (raiz quadrada do total de features) ou "log2".
>
> 4. min_samples_leaf (Mínimo de Amostras na Folha)
Estabelece o número mínimo de registros que um nó "final" (folha) deve conter. 
Uso: Ajuda a suavizar o modelo e reduzir o ruído. Se o valor for muito baixo (ex: 1), o modelo tende a ser muito específico; valores maiores tornam o modelo mais generalista.
>
> 5. min_samples_split (Mínimo para Divisão)
O número mínimo de dados necessários em um nó para que ele possa ser dividido em novos ramos. 
Uso: Funciona em conjunto com o max_depth para regularizar o crescimento das árvores. Valores como 2, 5 ou 10 são os pontos de partida mais frequentes
>
> 
>**Outros Parâmetros Operacionais Importantes**:
> 
>n_jobs: Defina como -1 para utilizar todo o poder de processamento do seu computador e acelerar o treino.
> 
>bootstrap: Quase sempre mantido como True (padrão) para permitir a amostragem com reposição.
