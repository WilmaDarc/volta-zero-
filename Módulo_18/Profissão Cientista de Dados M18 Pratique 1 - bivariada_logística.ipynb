Neste exerccio vamos trabalhar com a base "Heart Disease Data Set" disponvel no site do [UCI](https://archive.ics.uci.edu/ml/datasets/heart+disease).

Temos 303 indivduos e vamos utilizar 13 variveis disponveis para prever a presena de doena cardaca na varivel ```num```(a 14$^a$ varivel).

A descrio das variveis est recortada abaixo:
- age: idade do paciente em anos
- sex: sexo (1 = male; 0 = female)  
- cp: tipo de dor no peito
  - 1: angina tpica
  - 2: angina atpica
  - 3: dor no-angina
  - 4: assintomtico
- trestbps: presso sangunea em repouso (em mm Hg na admisso ao hospital
- chol: colesterol srico em mg/dl
- fbs: (acar no sangue em jejum > 120 mg/dl) (1 = True; 0 = False)
- restecg: resultados eletrocardiogrficos em repouso
  - 0: normal
  - 1: tendo anormalidade da onda ST-T (Inverses de onda T e / ou ST com elevao ou depresso de > 0.05 mV)
  - 2: mostrando hipertrofia ventricular esquerda provvel ou definitiva pelos critrios de Estes
- thalach: frequncia cardaca mxima alcanada
- exang: angina induzida por exerccio(1 = sim; 0 = no)
- oldpeak = Depresso de ST induzida por exerccio em relao ao repouso
- slope: Depresso de ST induzida por exerccio em relao ao repouso
  - 1: inclinao ascendente
  - 2: estvel
  - 3: inclinao descendente
- ca: nmero de vasos principais (0-3) coloridos por fluorosopia
- thal: 3 = normal; 6 = defeito corrigido; 7 = defeito reversvel
- num: diagnstico de doena cardaga (status de doena angiogrfica)

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import statsmodels.formula.api as smf
%matplotlib inline
````

````python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'

df = pd.read_csv(url, 
                 names=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
                        'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num'])
df.head()
````

````python
df['flag_doente'] = (df['num'] != 0).astype('int64')
df
````

1. Monte uma bivariada para a varivel ```sex``` com as seguintes estatsticas:

- Nmero de *saudveis*
- Nmero de *doentes*
- Total
- Quantidade mdia de doentes
- *Odds*
- *Odds ratio* (Use o total como referncia)
- Logito (LOG(*Odds*))
- O LOG do *Odds ratio* (com o total como referncia)  chamado de *Weight of Evidence* (WOE). Calcule-o.

````python
tab_sex = (pd.crosstab(index=df['sex'], 
                       columns=df['flag_doente'], 
                       margins=True, 
                       margins_name='total')
             .rename(index={0:'female', 1:'male'}, columns={0:'healthy', 1:'sick'}))

tab_sex['mean_healthy'] = tab_sex['healthy'] / tab_sex['total']
tab_sex['mean_sick'] = tab_sex['sick'] / tab_sex['total']

tab_sex['odds'] = tab_sex['sick'] / tab_sex['healthy']
tab_sex['odds_ratio'] = tab_sex['odds'] / tab_sex.loc['total', 'odds']

tab_sex['logito'] = np.log(tab_sex['odds'])
tab_sex['WOE'] = np.log(tab_sex['odds'] / tab_sex.loc['total', 'odds'])

tab_sex
````

2. Monte uma bivariada para a varivel ```age``` com as seguintes estatsticas:
- Categorize a varivel em 5 grupos
- Calcule as mesmas quantidades calculadas no execcio anterior
- Calcule tambm o valor mdio da varivel ```age```

````python
# Categorize a varivel em 5 grupos
df['faixa_idade'] = pd.qcut(df['age'], q=5)

# Criar a tabela de contingncia por faixa de idade
tab_age = pd.crosstab(index=df['faixa_idade'],
                      columns=df['flag_doente'],
                      margins=False).rename(columns={0: 'healthy', 1: 'sick'})

# Adicionar total por faixa
tab_age['total'] = tab_age['healthy'] + tab_age['sick']

# Calcular propores
tab_age['mean_healthy'] = tab_age['healthy'] / tab_age['total']
tab_age['mean_sick'] = tab_age['sick'] / tab_age['total']

# Calcular odds e odds ratio
tab_age['odds'] = tab_age['sick'] / tab_age['healthy']
odds_total = df['flag_doente'].sum() / (df.shape[0] - df['flag_doente'].sum())
tab_age['odds_ratio'] = tab_age['odds'] / odds_total

# Calcular logito e WOE
tab_age['logito'] = np.log(tab_age['odds'])
tab_age['WOE'] = np.log(tab_age['odds_ratio'])

# Calcular mdia da idade por faixa
tab_age['mean_age'] = df.groupby('faixa_idade')['age'].mean().round(2)

tab_age
````

3. Visualize a informao da bivariada construda acima atravs de um grfico da sua escolha.

````python
sns.set_theme()

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,4))

ax0 = tab_sex[['healthy', 'sick']].plot.bar(ax=axes[0],
color=['#4CAF50', '#F44336']) 
ax0.set_title(label='Nmero de saudveis X doentes')
ax0.set_xlabel(xlabel='Sexo')
ax0.set_xticklabels(labels=['Feminino', 'Masculino', 'Total'], rotation=0)
ax0.legend(labels=['Saudveis', 'Doentes'], loc=8, bbox_to_anchor=(0.5, -.4))


ax1 = tab_sex[['mean_healthy', 'mean_sick']].plot.bar(ax=axes[1], stacked=True,
                                                      color=['#4CAF50', '#F44336'])  
ax1.set_title(label='Probabilidade')
ax1.set_xlabel(xlabel='Sexo')
ax1.set_xticklabels(labels=['Feminino', 'Masculino', 'Total'], rotation=0)
ax1.legend(labels=['Mdia de saudveis', 'Mdia de doentes'], loc=8, bbox_to_anchor=(0.5, -.4))

ax2 = tab_sex['odds'].plot.bar(ax=axes[2],color=[ '#F44336']) 
ax2.set_title(label='Chance')
ax2.set_xlabel(xlabel='Sexo')
ax2.set_xticklabels(labels=['Feminino', 'Masculino', 'Total'], rotation=0)

plt.show()
````

````python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,4))

ax0 = tab_age.iloc[:-1,].plot.bar(x='mean_age', y=['healthy', 'sick'], ax=axes[0],color=['#4CAF50', '#F44336'])
ax0.set_title(label='Nmero de saudveis X doentes')
ax0.set_xlabel(xlabel='Idade')
ax0.set_xticklabels(labels=ax0.get_xticklabels(), rotation=0)
ax0.legend(labels=['Saudveis', 'Doentes'], loc=8, bbox_to_anchor=(0.5, -.4))

ax1 = tab_age.iloc[:-1].plot.bar(y=['mean_healthy', 'mean_sick'], ax=axes[1], stacked=True,color=['#4CAF50', '#F44336'])
ax1.set_title(label='Probabilidade')
ax1.set_xlabel(xlabel='Idade')
ax1.set_xticklabels(labels=ax0.get_xticklabels(), rotation=0)
ax1.legend(labels=['Mdia de saudveis', 'Mdia de doentes'], loc=8, bbox_to_anchor=(0.5, -.4))

ax2 = tab_age.iloc[:-1].plot.bar(y='odds', ax=axes[2],color=[ '#F44336'])
ax2.set_title(label='Chance')
ax2.set_xlabel(xlabel='Idade')
ax2.set_xticklabels(labels=ax0.get_xticklabels(), rotation=0)
ax2.legend(labels='')

plt.show()
````
