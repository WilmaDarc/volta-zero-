## Desenvolvimento do modelo

Desenvolva um modelo de *credit scoring* através de uma regressão logística.

- Trate valores missings e outliers
- Trate 'zeros estruturais'
- Faça agrupamentos de categorias conforme vimos em aula
- Proponha uma equação preditiva para 'mau'
- Caso hajam categorias não significantes, justifique

````python
def IV(variavel, resposta):
    tab = pd.crosstab(variavel, resposta, margins=True, margins_name='total')

    tilo_evento = tab.columns[0]
    tilo_nao_evento = tab.columns[1]

    tab['pct_evento'] = tab[tilo_evento]/tab.loc['total', tilo_evento]
    tab['ep'] = tab[tilo_evento]/tab.loc['total', tilo_evento]

    tab['pct_nao_evento'] = tab[tilo_nao_evento] / tab.loc['total', tilo_nao_evento]
    tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
    tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe
    return tab['iv_parcial'].sum()
````

````python
def biv_discreta(var, df):
    df['bom'] = 1-df.mau
    g = df.groupby(var)

    biv = pd.DataFrame({'qt_bom': g['bom'].sum(),
                        'qt_mau': g['mau'].sum(),
                        'mau': g['mau'].mean(),
                        var: g['mau'].mean().index,
                        'cont': g[var].count()})

    biv['ep'] = (biv.mau*(1-biv.mau)/biv.cont)**.5
    biv['mau_sup'] = biv.mau+t.ppf([0.975], biv.cont-1)*biv.ep
    biv['mau_inf'] = biv.mau+t.ppf([0.025], biv.cont-1)*biv.ep

    biv['logit'] = np.log(biv.mau/(1-biv.mau))
    biv['logit_sup'] = np.log(biv.mau_sup/(1-biv.mau_sup))
    biv['logit_inf'] = np.log(biv.mau_inf/(1-biv.mau_inf))

    tx_mau_geral = df.mau.mean()
    woe_geral = np.log(df.mau.mean() / (1 - df.mau.mean()))

    biv['woe'] = biv.logit - woe_geral
    biv['woe_sup'] = biv.logit_sup - woe_geral
    biv['woe_inf'] = biv.logit_inf - woe_geral

    fig, ax = plt.subplots(2, 1, figsize=(8, 6))
    ax[0].plot(biv[var], biv.woe, ':bo', label='woe')
    ax[0].plot(biv[var], biv.woe_sup, 'o:r', label='limite superior')
    ax[0].plot(biv[var], biv.woe_inf, 'o:r', label='limite inferior')

    num_cat = biv.shape[0]
    ax[0].set_xlim([-.3, num_cat-.7])
    ax[0].set_ylabel("Weight of Evidence")
    ax[0].legend(bbox_to_anchor=(.83, 1.17), ncol=3)

    ax[0].set_xticks(list(range(num_cat)))
    ax[0].set_xticklabels(biv[var], rotation=8)

    ax[1] = biv.cont.plot.bar()
    return biv
````

````python
def biv_continua(var, ncat, df):
    df['bom'] = 1-df.mau
    cat_srs, bins = pd.qcut(df[var], ncat, retbins=True, precision=0, duplicates='drop')
    g = df.groupby(cat_srs)

    biv = pd.DataFrame({'qt_bom': g['bom'].sum(),
                        'qt_mau': g['mau'].sum(),
                        'mau': g['mau'].mean(),
                        var: g[var].mean(),
                        'cont': g[var].count()})

    biv['ep'] = (biv.mau*(1-biv.mau)/biv.cont)**.5
    biv['mau_sup'] = biv.mau+t.ppf([0.975], biv.cont-1)*biv.ep
    biv['mau_inf'] = biv.mau+t.ppf([0.025], biv.cont-1)*biv.ep

    biv['logit'] = np.log(biv.mau/(1-biv.mau))
    biv['logit_sup'] = np.log(biv.mau_sup/(1-biv.mau_sup))
    biv['logit_inf'] = np.log(biv.mau_inf/(1-biv.mau_inf))

    tx_mau_geral = df.mau.mean()
    woe_geral = np.log(df.mau.mean() / (1 - df.mau.mean()))

    biv['woe'] = biv.logit - woe_geral
    biv['woe_sup'] = biv.logit_sup - woe_geral
    biv['woe_inf'] = biv.logit_inf - woe_geral

    fig, ax = plt.subplots(2, 1, figsize=(8, 6))
    ax[0].plot(biv[var], biv.woe, ':bo', label='woe')
    ax[0].plot(biv[var], biv.woe_sup, 'o:r', label='limite superior')
    ax[0].plot(biv[var], biv.woe_inf, 'o:r', label='limite inferior')

    num_cat = biv.shape[0]
    ax[0].set_ylabel("Weight of Evidence")
    ax[0].legend(bbox_to_anchor=(.83, 1.17), ncol=3)

    ax[1] = biv.cont.plot.bar()
    return None
````

### Pipeline

Crie um pipeline contendo essas funções.

preprocessamento()
- substituicao de nulos
- remoção outliers
- PCA
- Criação de dummy de pelo menos 1 variável (posse_de_veiculo)

````python
# --- 1. Definição das Colunas Relevantes para o Pipeline ---
# Variáveis numéricas para imputação de nulos e escalonamento
numerical_features_for_pipeline = ['tempo_emprego', 'idade', 'qt_pessoas_residencia']
# Variável 'renda' para transformação logarítmica (para lidar com outliers) e escalonamento
renda_feature_for_pipeline = ['renda']
# Variáveis categóricas para criação de dummies (incluindo posse_de_veiculo)
categorical_features_for_pipeline = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']

# --- 2. Transformador Customizado para Agrupamento de Categorias ---
# Classe para aplicar agrupamentos de categorias de forma consistente dentro do pipeline
class CategoricalGrouper(BaseEstimator, TransformerMixin):
    def __init__(self, mappings):
        self.mappings = mappings

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X_copy = X.copy() # Trabalha em uma cópia para evitar modificar o dataframe original
        for col, map_dict in self.mappings.items():
            if col in X_copy.columns:
                X_copy[col] = X_copy[col].replace(map_dict)
        return X_copy

# Mapeamentos para agrupamento de categorias (conforme análises anteriores)
categorical_grouping_mappings = {
    'educacao': {
        'Fundamental': 'Fund/Médio/Sup Incomp',
        'Médio': 'Fund/Médio/Sup Incomp',
        'Superior incompleto': 'Fund/Médio/Sup Incomp',
        'Superior completo': 'Sup Comp/Pós',
        'Pós graduação': 'Sup Comp/Pós'
    },
    'tipo_renda': {
        'Assalariado': 'Assalariado/Bolsista',
        'Bolsista': 'Assalariado/Bolsista'
    }
}

# --- 3. Pipelines Específicos para Tipos de Variáveis ---

# Pipeline para variáveis numéricas padrão (substituição de nulos com a média e escalonamento)
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')), # Substituição de nulos
    ('scaler', StandardScaler()) # Escalonamento
])

# Pipeline para a variável 'renda' (substituição de nulos, transformação logarítmica e escalonamento)
# A transformação logarítmica (np.log1p) ajuda na remoção/mitigação de outliers e assimetria
renda_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')), # Substituição de nulos (se houver)
    ('log_transform', FunctionTransformer(np.log1p, validate=True)), # Remoção de outliers / tratamento de assimetria
    ('scaler', StandardScaler()) # Escalonamento
])

# Pipeline para variáveis categóricas (agrupamento e criação de dummies)
categorical_pipeline = Pipeline(steps=[
    ('grouper', CategoricalGrouper(mappings=categorical_grouping_mappings)), # Agrupamento de categorias
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first')) # Criação de dummy
])

# --- 4. ColumnTransformer para Combinar os Pipelines ---
# Combina os transformadores específicos para cada tipo de coluna
preprocessor_combiner = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features_for_pipeline),
        ('renda_proc', renda_pipeline, renda_feature_for_pipeline),
        ('cat', categorical_pipeline, categorical_features_for_pipeline)
    ],
    remainder='passthrough' # Mantém outras colunas que não foram especificadas
)

# --- 5. Pipeline Completo de Pré-processamento Final ---
# O pipeline final que aplica todos os passos anteriores e o PCA
final_preprocessing_pipeline = Pipeline(steps=[
    ('preprocessor_combiner', preprocessor_combiner),
    ('pca', PCA(n_components=5)) # PCA para redução de dimensionalidade
])

print("Pipeline de pré-processamento 'final_preprocessing_pipeline' criado com sucesso, contendo nulos, outliers, PCA e dummies.")

# Exemplo de como você aplicaria o pipeline (opcional e comentado)
# import pandas as pd
# # Recarregue o df_train_initial como estava antes das transformações diretas para aplicar o pipeline nele
# df_original_raw = pd.read_feather("credit_scoring.ftr")
# df_original_raw.set_index('data_ref', inplace=True)
# df_train_raw = df_original_raw['2015-01-01':'2015-12-01'].reset_index()
# df_train_raw['mau'] = df_train_raw.mau.astype('int64')
# df_train_raw.drop(columns=['data_ref', 'index', 'mau', 'bom', 'score'], errors='ignore', inplace=True)

# X_train_processed_by_new_pipeline = final_preprocessing_pipeline.fit_transform(df_train_raw)
# print(f"\nShape do X_train processado pelo novo pipeline: {X_train_processed_by_new_pipeline.shape}")
````

### 3. Avaliar o Modelo e Gerar Gráficos Finais

Vamos calcular as métricas de acurácia, AUC, Gini e KS para o conjunto de treino e teste, e também gerar os gráficos solicitados.

````python
def evaluate_model_and_plot(model, X, y, dataset_name):
    # Previsões de probabilidade
    y_pred_proba = model.predict_proba(X)[:, 1]

    # Previsões de classe (usando 0.5 como threshold padrão para acurácia)
    y_pred_class = (y_pred_proba > 0.5).astype(int)

    # Acurácia
    acc = accuracy_score(y, y_pred_class)

    # AUC
    fpr, tpr, thresholds = roc_curve(y, y_pred_proba)
    auc_score = roc_auc_score(y, y_pred_proba)

    # Gini
    gini = 2 * auc_score - 1

    # KS
    # Verifica se há amostras para ambas as classes (bom e mau) para calcular o KS
    if np.sum(y == 1) > 0 and np.sum(y == 0) > 0:
        ks = ks_2samp(y_pred_proba[y == 1], y_pred_proba[y == 0]).statistic
    else:
        ks = np.nan # Não é possível calcular KS se uma das classes estiver vazia

    print(f"\n--- Métricas de Avaliação ({dataset_name}) ---")
    print(f"Acurácia: {acc:.1%}")
    print(f"AUC: {auc_score:.1%}")
    print(f"GINI: {gini:.1%}")
    if not np.isnan(ks):
        print(f"KS: {ks:.1%}")
    else:
        print("KS: Não foi possível calcular (faltam amostras de uma das classes).")

    # Plot ROC curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {auc_score:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Curva ROC ({dataset_name})')
    plt.legend(loc="lower right")
    plt.show()

    # Plot KS curve
    plt.figure(figsize=(8, 6))

    # Para evitar erros se uma das classes não tiver amostras
    if np.sum(y == 1) > 0:
        plt.plot(np.sort(y_pred_proba[y == 1]), np.linspace(0, 1, len(y_pred_proba[y == 1]), endpoint=False), label='Mau')
    if np.sum(y == 0) > 0:
        plt.plot(np.sort(y_pred_proba[y == 0]), np.linspace(0, 1, len(y_pred_proba[y == 0]), endpoint=False), label='Bom')

    plt.title(f'Curva KS ({dataset_name})')
    plt.xlabel('Probabilidade')
    plt.ylabel('Função de Distribuição Acumulada')
    plt.legend()
    plt.show()

# Avaliar no conjunto de treino
evaluate_model_and_plot(lgbm_model, X_train_processed_lgbm, y_train_lgbm, 'Treino')

# Avaliar no conjunto de teste
evaluate_model_and_plot(lgbm_model, X_test_processed_lgbm, y_test_lgbm, 'Teste')
````
