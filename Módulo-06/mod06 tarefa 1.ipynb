#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv("Profissão Cientista de Dados M6 - SINASC_RO_2019.csv")

# Visualizar as primeiras linhas
df.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de variáveis
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

# Criar gráficos de barras
for var in variaveis:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=var, order=df[var].value_counts().index)
    plt.title(f'Frequência de {var}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Calcular média de APGAR5 por número de consultas
media_apgar5 = df.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# Plotar gráfico
plt.figure(figsize=(10, 5))
sns.barplot(data=media_apgar5, x='CONSULTAS', y='APGAR5')
plt.title('APGAR5 médio por número de CONSULTAS')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
variaveis_painel = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

for ax, var in zip(axes.flatten(), variaveis_painel):
    sns.countplot(data=df, x=var, order=df[var].value_counts().index, ax=ax)
    ax.set_title(f'Frequência de {var}')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
