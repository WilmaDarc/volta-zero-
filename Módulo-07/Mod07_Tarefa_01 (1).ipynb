1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.

````python
import numpy as np
from scipy.stats import chi2_contingency
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

# 1) carregar a base
fname = "Profissão Cientista de Dados M7 A1 - demo01.csv"  # ajuste caminho se necessário
df = pd.read_csv(fname)
print("Arquivo carregado. Dimensão:", df.shape)
display(df.head())

# 1b) inspecionar colunas e tipos
print("Colunas:", df.columns.tolist())
display(df.info())

# 2) montar metadados
meta = pd.DataFrame({
    'variable': df.columns,
    'dtype': [str(df[col].dtype) for col in df.columns],
    'n_unique': [df[col].nunique(dropna=False) for col in df.columns],
    'n_missing': [df[col].isna().sum() for col in df.columns]
})

def var_type(col_series):
    if pd.api.types.is_numeric_dtype(col_series):
        return 'numérica'
    else:
        return 'categórica'

meta['type'] = [var_type(df[c]) for c in meta['variable']]

def papel(c):
    if c == 'renda':
        return 'variável resposta (renda)'
    if c == 'mau':
        return 'variável resposta (mau)'
    return 'explicativa'

meta['papel_no_modelo'] = [papel(c) for c in meta['variable']]

# adicionar qt_cat_categories (renomeando n_unique)
meta = meta.rename(columns={'n_unique': 'qt_cat_categories'})
display(meta)

# 3) criar dummies e preparar dataframe para scikit-learn
# remover colunas do tipo string que sejam IDs/texto longos (ajuste to_drop se necessário)
to_drop = []  # se tiver 'nome' ou 'id', adicione aqui: ['id', 'nome']
df_model = df.drop(columns=[c for c in to_drop if c in df.columns])

# detectar colunas categóricas
cat_cols = [c for c in df_model.columns if df_model[c].dtype == 'object' or df_model[c].dtype.name == 'category']
print("Colunas categóricas detectadas:", cat_cols)

# criar dummies (drop_first=True para evitar multicolinearidade)
df_dummies = pd.get_dummies(df_model, columns=cat_cols, drop_first=True)

# garantir que 'renda' e 'mau' são numéricas quando presentes
if 'renda' in df_dummies.columns and df_dummies['renda'].dtype == 'object':
    df_dummies['renda'] = pd.to_numeric(df_dummies['renda'].astype(str).str.replace(',','.'), errors='coerce')
if 'mau' in df_dummies.columns and df_dummies['mau'].dtype == 'object':
    df_dummies['mau'] = df_dummies['mau'].astype(str).str.lower().map({'sim':1,'s':1,'yes':1,'1':1,'true':1,'verdadeiro':1,'não':0,'nao':0,'n':0,'no':0,'0':0,'false':0}).astype(float)

print("Dataframe com dummies pronto. Dimensão:", df_dummies.shape)
display(df_dummies.head())

# 4) comparar posse_de_veiculo e posse_de_imovel para prever mau = 1
c1 = 'posse_de_veiculo'
c2 = 'posse_de_imovel'
target = 'mau'

# segurança: verificar se colunas existem
for col in [c1, c2, target]:
    if col not in df.columns and col not in df_dummies.columns:
        print(f"Aviso: coluna '{col}' não encontrada no dataset. Colunas disponíveis: {df.columns.tolist()}")
        
# tentar trabalhar com as versões em df (originais) quando possível
df_for_analysis = df.copy()

# converter colunas de posse e target para binário caso estejam em texto
def to_binary(series):
    if pd.api.types.is_numeric_dtype(series):
        return series
    s = series.astype(str).str.lower()
    return s.map({'sim':1,'s':1,'yes':1,'y':1,'1':1,'true':1,'t':1,'não':0,'nao':0,'n':0,'no':0,'0':0}).astype(float)

if c1 in df_for_analysis.columns:
    df_for_analysis[c1] = to_binary(df_for_analysis[c1])
if c2 in df_for_analysis.columns:
    df_for_analysis[c2] = to_binary(df_for_analysis[c2])
if target in df_for_analysis.columns:
    df_for_analysis[target] = to_binary(df_for_analysis[target])

# tabela cruzada entre c1 e c2 (se existirem)
if c1 in df_for_analysis.columns and c2 in df_for_analysis.columns:
    ct = pd.crosstab(df_for_analysis[c1], df_for_analysis[c2], margins=True)
    print("Tabela cruzada {} x {}:".format(c1, c2))
    display(ct)

# taxas de mau por categoria
def taxa_mau_por(col):
    if col not in df_for_analysis.columns or target not in df_for_analysis.columns:
        return pd.DataFrame()
    tab = df_for_analysis.groupby(col)[target].agg(['mean','count']).rename(columns={'mean':'taxa_mau','count':'n'})
    return tab.sort_values('taxa_mau', ascending=False)

print("Taxa mau por", c1)
display(taxa_mau_por(c1))
print("Taxa mau por", c2)
display(taxa_mau_por(c2))

# teste qui-quadrado (se possível)
if c1 in df_for_analysis.columns and target in df_for_analysis.columns:
    cont1 = pd.crosstab(df_for_analysis[c1], df_for_analysis[target])
    if cont1.size and cont1.shape[0] > 1 and cont1.shape[1] > 1:
        chi1 = chi2_contingency(cont1)
        print("Qui-quadrado {} vs {}: stat={:.4f}, p-value={:.4e}".format(c1, target, chi1[0], chi1[1]))
    else:
        print("Não é possível aplicar qui-quadrado para", c1)

if c2 in df_for_analysis.columns and target in df_for_analysis.columns:
    cont2 = pd.crosstab(df_for_analysis[c2], df_for_analysis[target])
    if cont2.size and cont2.shape[0] > 1 and cont2.shape[1] > 1:
        chi2r = chi2_contingency(cont2)
        print("Qui-quadrado {} vs {}: stat={:.4f}, p-value={:.4e}".format(c2, target, chi2r[0], chi2r[1]))
    else:
        print("Não é possível aplicar qui-quadrado para", c2)

# 5) AUC univariada para comparar poder preditivo (somente se target existir e for binário)
if target in df_dummies.columns:
    y = df_dummies[target].dropna()
    results = {}
    for col in [c1, c2]:
        # tentar encontrar colunas correspondentes em df_dummies (por exposição direta ou dummies)
        if col in df_dummies.columns:
            Xcol = df_dummies[[col]].fillna(0)
        else:
            # procurar dummies que comecem com o nome da coluna
            matched = [c for c in df_dummies.columns if c.startswith(col + "_")]
            if matched:
                Xcol = df_dummies[matched].fillna(0)
            else:
                print(f"Não encontrei variáveis para '{col}' em df_dummies. Pulando AUC para essa variável.")
                continue
        try:
            clf = LogisticRegression(max_iter=1000)
            auc = cross_val_score(clf, Xcol, y, cv=5, scoring='roc_auc', error_score='raise').mean()
            results[col] = auc
        except Exception as e:
            print("Erro ao calcular AUC para", col, ":", e)
    print("AUC univariada (maior = melhor):", results)
else:
    print("Coluna target 'mau' não está disponível em df_dummies para cálculo de AUC.")

# 6) salvar base preparada
outname = "demo01_preparada.csv"
df_dummies.to_csv(outname, index=False)
print("Base preparada salva em:", outname)
````
