```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv("Profissão Cientista de Dados M6 - SINASC_RO_2019.csv")

# Visualizar as primeiras linhas
df.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='IDADEMAE')
plt.title('Boxplot da Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='PESO')
plt.title('Boxplot do Peso do Bebê')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='IDADEMAE', bins=20, kde=True)
plt.title('Histograma da Idade da Mãe')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='PESO', bins=20, kde=True)
plt.title('Histograma do Peso do Bebê')
plt.show()
```
